```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder , OneHotEncoder
```

```python
# Let's say 'df' is your original dataset
# Replace 'df' with your actual dataset name

# Columns to ordinal encode
ordinal_cols = [ 'Adaptability', 'TimeManagement', 'Curiosity',
       'Integration', 'Versatality', 'Scheduling', 'Efficiency',
       'Multitasking', 'Consistency', 'Participation', 'ExtraActivities',
       'Engagement', 'Collaboration', 'Balance', 'Cooperation',
       'ConflictResolution', 'Synergy', 'Empathy', 'Resilience', 'Feedback',
       'SelfCare', 'Resourcefulness', 'FinancialConstraint', 'Wellness',
       'Stability']

# Column to nominal encode
nominal_col = ['Hardship']

# Create a new DataFrame for encoding
encoded_df = df.copy()

# Ordinal encoding
ordinal_encoder = OrdinalEncoder()
encoded_df[ordinal_cols] = ordinal_encoder.fit_transform(df[ordinal_cols])

# Nominal encoding (One-Hot Encoding)
one_hot_encoder = OneHotEncoder(sparse_output=False, drop='first')  # Adjusted 'sparse' to 'sparse_output'
nominal_encoded = one_hot_encoder.fit_transform(df[nominal_col])
nominal_encoded_df = pd.DataFrame(
    data=nominal_encoded,
    columns=[f'{nominal_col[0]}_{i}' for i in range(nominal_encoded.shape[1])]
)

# Concatenate nominal encoded columns with the encoded DataFrame
encoded_df = pd.concat([encoded_df, nominal_encoded_df], axis=1)

# Now 'encoded_df' contains the encoded columns along with the original ones
# You can save this new dataset to a new file if you'd like
save_path= r'C:/Users/Incredible-PC/Documents/encoded_dataset1.csv'
encoded_df.to_csv(save_path, index=False)  # Change the file name as needed
print("Saved")


```
